```python
time_period_data = a1.get_time_periods_dfs(data)
listening_history = a1.artist_level_data(relevant_data, time_period_data, 0)
listening_history = listening_history.reset_index(level = 0) \
                        .rename(columns ={"level_0" : "artistName"})
listening_history['date'] = listening_history.index
listening_history = listening_history.reset_index()
for i in listening_history.index:
    listening_history.loc[i, 'date'] = listening_history.loc[i, 'date'].date()
listening_history = listening_history.rename(columns = {"artistName" : "Artist", 
                                                        "sum" : "Total Listens", 
                                                        "week" : "Week",
                                                        "prop" : "Proportion",
                                                        "date" : "Date"})
```

```python
caption = "Fig 2: Top 10 most listened to artists in this dataset, animated over the time frame March 2019 - March 2020"
idx = relevant_data.groupby(['artistName'])['total_listens'].transform(max) == relevant_data['total_listens']
relevant_sum = relevant_data[idx]
ordered_artists = list(relevant_sum.artistName)
fig = px.bar(listening_history, x="Artist", y="Total Listens", 
              animation_frame="Week", animation_group="Artist", range_y=[0,600],
              category_orders={'Artist':ordered_artists}, opacity = 0.6)
fig.update_layout(title_text='Top 10 Listened to Artists 2019-2020 - Animation',
                    xaxis=dict(title_text="Artist"),
                    yaxis=dict(title_text="Songs Played"),
                    margin_b=220)
fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 250
fig.show()
```

```python
def get_caption(caption):
    return dict(xref='paper',
            yref='paper',
            x=0.5, y=-0.88,
            showarrow=False,
            text = caption)
```

```python
caption = "Fig 9: Breakdown, by artist, of all the songs listened to in <br> Week 1 (March 9, 2019 - March 15, 2019). Artists that make up at least 50% of <br> weekly listening are pulled out." 
time_period_data = a1.get_time_periods_dfs(data)
for d in time_period_data.values():
    props = d['props']
    props['Artist'] = props.index
    props = props.sort_values(by = 'proportion', ascending = False)
    half_prop = 0
    pull_arr = np.zeros(len(props.index))
    count_pull = 0
    while half_prop < 0.5:
        p = props.iloc[count_pull]['proportion']
        half_prop += p
        pull_arr[count_pull] = 0.25
        count_pull += 1
    fig = go.Figure(go.Pie(labels=props['Artist'], values=props['count'], pull = pull_arr))
    fig.update_traces(textposition='inside', textinfo='percent+label')
    fig.update_layout(uniformtext_minsize=14, uniformtext_mode='hide')
    fig.update_layout(
        autosize=False,
        width=800,
        height=500,
        margin=dict(
            l=50,
            r=50,
            b=100,
            t=100,
            pad=4
        )
    )
    fig.update_traces(hoverinfo='label+percent', textinfo='label', textfont_size=20,
                  marker=dict(line=dict(color='#000000', width=1)))
    fig.update_layout(title_text="Songs Listened to in Week 1: Divided by Artist",
                        margin_b=100, annotations = [dict(xref='paper',
                                        yref='paper',
                                        x=0.5, y=-0.3,
                                        showarrow=False,
                                        text = caption)])
    fig.update_layout(title_text="Songs Listened to in Week 1: Divided by Artist")
    fig.show()
    break
py.plot(fig, filename = 'one_pie_half_pull', auto_open=False)
```
